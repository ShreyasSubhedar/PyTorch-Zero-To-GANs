## Function 2 - torch.sigmoid

It returns the sigmoid neuron of given tensor, sigmoid neuron is a popular model of classification in Deep Learning.

```python
# Example 1 
w =  torch.randn(10)
x = torch.randn(10)
b = torch.rand(1)
perceptron = x@w.t()+b        #perceptron model
print(perceptron)
print("Sigmoid for the given perceptron")
print(torch.sigmoid(perceptron))
```

Here I use some random data for weights, inputs and bias anf first calculate the perceptron and then pass this value to the sigmoid.

```python
# Example 2 - 

percetron = torch.tensor([0.0])
print(torch.sigmoid(perceptron))
```

As we know the value of sigmoid is 1/2 (i.e 0.5) when `w*x+b` is zero.

```python
# Example 3 - breaking (to illustrate when it breaks)
print(torch.sigmoid(torch.tensor([0])))
```

Passing long value as a tensor in sigmoid gives us an error message `"sigmoid_cpu" not implemented for 'Long'`

Sigmoid function is widely use for traing the model and parameter, it is heavily used for DNN.
